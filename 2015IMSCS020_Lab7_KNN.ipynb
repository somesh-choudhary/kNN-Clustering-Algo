```python
def distance(x,y):
    d = 0
    for i in range(len(x)):
        d += (x[i] - y[i])**2
    d = np.sqrt(d)
    return d
```

```python
p = np.array([5. , 3.6, 1.4, 0.2])
```

```python
def distinct(lst):
    x = []
    for i in lst:
        if i not in x:
            x.append(i)
    return x
```

```python
def KNN(X,y,p,k=3):
    lst = [99999999999]*len(X)
    neighbour = [999999]*k
    for i in range(len(X)):
        lst[i] = (distance(X[i],p))
    lst = np.argsort(lst)
    for i in range(k):
        neighbour[i] = (y[lst[i]])
    print(neighbour)
    distinct_neighbour = distinct(neighbour)
    ans = distinct_neighbour[0]
    count = 0 
    for i in (distinct_neighbour):
        c = neighbour.count(i)
        if c > count:
            count = c
            ans = i
    return ans
```

```python
KNN(X,y,p,k=100)
```
